```python
import pandas as pd

# Cargar los archivos Excel en Pandas con el motor openpyxl
df_inventario = pd.read_excel("data/INVENTARIO.xlsx", engine="openpyxl")
df_clientes = pd.read_excel("data/CLIENTES.xlsx", engine="openpyxl")
df_ventas = pd.read_excel("data/VENTAS.xlsx", engine="openpyxl")
```

```python
# Mostrar los primeros 10 registros de cada dataset en forma de tabla

from IPython.display import display

print("Inventario:")
display(df_inventario.head(10))

print("Clientes:")
display(df_clientes.head(10))

print("Ventas:")
display(df_ventas.head(10))
```

Exploracion y limpieza de datos

1. Exploraciòn de los datos

Revisiòn de la estructura y tipo de datos

```python
# Información general de cada dataset

print("\n Información del Inventario:")
display(df_inventario.info())

print("\n Información de Clientes:")
display(df_clientes.info())

print("\n Información de Ventas:")
display(df_ventas.info())
```

Validaciòn de valores nulos en cada base de datos

```python
# Contar valores nulos en cada dataset

print("\n Valores nulos en Inventario:")
display(df_inventario.isnull().sum())

print("\n Valores nulos en Clientes:")
display(df_clientes.isnull().sum())

print("\n Valores nulos en Ventas:")
display(df_ventas.isnull().sum())
```

Revisiòn de valores ùnicos en columnas clave

```python
# Mostrar valores únicos en columnas categóricas

print("\n Categorías únicas en Inventario:")
for col in df_inventario.select_dtypes(include=['object']).columns:
    print(f"{col}: {df_inventario[col].unique()}")

print("\n Categorías únicas en Clientes:")
for col in df_clientes.select_dtypes(include=['object']).columns:
    print(f"{col}: {df_clientes[col].unique()}")

print("\n Categorías únicas en Ventas:")
for col in df_ventas.select_dtypes(include=['object']).columns:
    print(f"{col}: {df_ventas[col].unique()}")
```
